# Classifying 5x5 mazes - Using symmetry
In this notebook we use the same Quantum Machine Learing approach as for the Tic-Tac-Toe boards, but for a different application: determining whether a path exists between any two corners of a 5x5 maze. The following cells show some examples, where the mazes are colored green or red depending on whether it is solvable or not, respectively.

```python
from qiskit import *
import numpy as np
import bigmaze as m

from qiskit.quantum_info import SparsePauliOp
from qiskit.primitives import Estimator

def make_circuit(x, params, layers):
    circ = QuantumCircuit(25)
    for i in range(layers):
        l = i*21
        m.encode_data(x, circ)
        m.add_single_qubit_gates(params[l:l+12], circ)
        m.add_two_qubit_gates(params[l+12:l+21], circ)

    return circ
# Create a Quantum Circuit acting on a quantum register of nine qubits

layers = 1

n_params = 21
params = np.random.rand(n_params*layers)*2*np.pi
circ = make_circuit([1,1,0,0,0,1,0,0,0], params, layers)

maze_fields, expected_results = m.generate_maze_field(200)
```

```python
# Define the loss function
def l2_loss(output, target):
    output, target = np.array(output), np.array(target)
    return np.abs(output - target)**2
```

```python
x = maze_fields
y = expected_results
#print(x)
#print(y)

# shuffle the indices
shuffle_indices = np.random.permutation(len(x))
train_size = int(len(x) * 0.3)

# split the indices into training and testing sets
train_indices = np.array(shuffle_indices[:train_size])
test_indices = np.array(shuffle_indices[train_size:])

# create the training and testing sets
x_train, y_train = np.take(x, train_indices, axis=0), np.take(y, train_indices, axis=0)
x_test, y_test = np.take(x, test_indices, axis=0), np.take(y, test_indices, axis=0)


print("Example train data: ", x_train[17], y_train[17])
print(len(x))
```

```python
from tqdm import tqdm
# Define the optimizer
from qiskit.algorithms.optimizers import SPSA
optimizer = SPSA(maxiter = 100)

# Define the cost function to be minimized by the optimizer
def cost_function(params):

    cost = 0
    estimator = Estimator()
    observables = (
        SparsePauliOp("IIIIIIIIIIIIZIIIIIIIIIIII")
    )
    
    for x, y in tqdm(zip(x_train, y_train)):
        # Create a Quantum Circuit acting on a quantum register of nine qubits
        circ = make_circuit(x, params, layers)        

        circuits = (
            circ,
        )

        job = estimator.run(circuits, observables)
        result = job.result().values[0]
        
        prediction = (result + 1)/2
        
        cost += l2_loss(prediction, y)
        
    print(cost)
    return cost

# Initialize the parameters
params = np.random.rand(n_params*layers)*2*np.pi

# Train the circuit
print('Initial parameters:', params)

# Check the qiskit docs to figure out how to start an optimizer
result = optimizer.minimize(cost_function, x0 = params)
print("Finished optimization.")
```
